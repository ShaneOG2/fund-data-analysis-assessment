Adapt the code from 01-information Jupyer Notebook in notes to generate a 1000 character long string with weights based on the previous two characters.

Firstly, we start by getting the Alice in Wonderland book and storing off www.gutenberg.org, then store it as one big string called **sbook**.

```python
# Make HTTP requests for internet reqources.
import urllib.request

# For doing (pseudo-)random things in Python.
import random

# The URL of a text version of Alice in Wonderland.
book_url = 'https://www.gutenberg.org/files/11/11-0.txt'

# Get the book.
book = list(urllib.request.urlopen(book_url))

# Decode the lines and strip line endings.
book = [line.decode('utf-8-sig').strip() for line in book]

# Get the whole book in one big string.
sbook = ''.join(book[26:]).lower()
```

Secondly, we create a dictionary with the weighting system. 

The first key is each of the two previous characters possibilites (e.g. "aa", "ba", "ca",............,"y ", "z ", "  ") and it's value is another dictionary. The key of this dictionary is chars (each letter of the alphabet plus space) and its values are the weights of the key.  

```python
# All letters and a space.
chars = 'abcdefghijklmnopqrstuvwxyz '

# Create the weights.
threeghts = {(x + y): {z: sbook.count(x + y + z) for z in chars} for y in chars for x in chars}
threeghts
```

We have to start with two letters, so let's find the most common sequence of two letters.

Using Ian's code, convert into dictionary where the key is the combination of two chars and value is the weights. Order this to find most common. 

```python
# Create the weights.
twoghts = {c: {d: sbook.count(c + d) for d in chars} for c in chars}

counts = []
twos = []
for i in chars:
    for j in chars:
        counts.append(twoghts[i][j])
        twos.append(i + j)
twos = dict(zip(twos, counts))

# https://stackoverflow.com/a/613218
dict(sorted(twos.items(), key=lambda item: item[1], reverse=True))
```

"e ", "t " and "he" are the most common sequence occurances. This is followed by "th" which i choose to start with because i thought it would be more likely to be at the beginning of a piece of text.   

```python
# Start with "th".
pairs = 'th'
N = 1000
# Do the following N times.
for i in range(1, N-1):
    # Get the weights where the previous two character is the last two characters in pairs.
    wt = threeghts[pairs[-2:]]
    # Turn wt into a list, ordered by chars. 
    wt = [wt[c] for c in chars]
    # Randomly pick the next character using those weights.
    nextc = random.choices(chars, weights=wt, k=1)[0]
    # Append the character to twos.
    pairs = pairs + nextc
```

Taking the example below we can see the relationship between logs and powers from the image above:

```python
b = 2
x = 8
y = 3
```

```python
import math
y == math.log(x, b)
```

```python
b ** y == x
```

For the $log_2({8})$ we are asking how many 2s we need to multiply together to get 8? 

The answer is 3

```python
2 * 2 * 2 
```

Say the question is how many 2s we need to get 0? 

Our log would be $log_2({0})$ but this gives us an error. 

```python
math.log(0, 2)
```

Why? 

There is no way of getting 0 from $2^{x}$ where x is any rational number. 

Similarly, there is no way of getting a number less than 0 so x must be always greater than 0. 

See graph for $log_2({z})$ below where x = $2^z$. 

```python
import numpy as np
# Get a list of numbers between -10 and 4
z = np.arange(-10, 5)
# Get x = 2^z
two_power_z = np.power(2, z, dtype=float)
```

For x is between -10 and 4 get x = $2^{z}$. 

Numpy gives the answers in decimals but these are the values for x $2^\frac{1}{1024}$, $2^\frac{1}{512}$, $2^\frac{1}{256}$, $2^\frac{1}{128}$, $2^\frac{1}{64}$, $2^\frac{1}{32}$, $2^\frac{1}{16}$, $2^\frac{1}{8}$, $2^\frac{1}{4}$, $2^\frac{1}{2}$, 1, 2, 4, 8, and 16. 

To get the y values we need to get y = $log_2({x})$

```python
log_of_two_power_z = np.log2(two_power_z)
log_of_two_power_z
```

Therefore our (x, y) co-ordinates are: ($2^\frac{1}{1024}$, -10), ($2^\frac{1}{512}$, -9), ($2^\frac{1}{256}$, -8), ($2^\frac{1}{128}$, -7), ($2^\frac{1}{64}$, -6), ($2^\frac{1}{32}$, -5), ($2^\frac{1}{16}$, -4), ($2^\frac{1}{8}$, -3), ($2^\frac{1}{4}$, -2), ($2^\frac{1}{2}$, -1), (1, 0), (2, 1), (4, 2), (8, 3), and (16, 4). 

```python
# I want to plot a line x = 0 so zeros are my x co-ordinates and i'll use the same y co-ordinates as I did for log2(x)
zeros = log_of_two_power_z * 0
zeros
```
