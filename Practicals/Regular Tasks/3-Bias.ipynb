Give three real-world examples of different types of cognitive bias.

**Cognitive bias** can be defined as a set of predictable mental errors that arise from our limited ability to process information objectively. It can result in illogical and irrational decisions, and it can cause you to misjudge risks and threats.

There are many types of cognitive bias. Here are three types and examples:
- **Anchoring** - The tendency to jump to conclusions - to base your final decision on information you gained early in the the decision making process. 
- **Confirmation** - You look for information that supports your existing beliefs and reject data that goes against what you believe. This can lead to biased decisions which don't factor in all information. 
- **Gambler's Fallacy** - You expect past events to influence the future. 

**Examples:**
- **Anchoring** - Anchors often effect decision making when it comes to purchasing a product. For example, you are looking to buy a new car. You go to a dealership and you see one you like. The salesperson says it is €30,000. This is too expensive for you so you keep looking around. A week later the salesperson rings and say they can sell it for €25,000. You immediately agree to it. You feel like you are getting a good deal when they lowered the price from €30,000 to €25,000, even if the real value of the car is less than €25,000. The first offer you heard determined the value of the car, rather than its age, miles and quality.
- **Confirmation** - Watching a game of football when the team you are supporting are playing. You tend to think a referee made a good call when it is beneficial for your team, but if it goes against your team, there’s a good chance you will think the referee made a bad call. After the match you are on the losing side and can come up with a number of calls that went against your team. Fans on the opposing side can do the exact same thing in support of their team’s victory. 
- **Gambler's Fallacy** - If a couple has 3 boys and are now trying for a girl, they might think that because they have 3 boys already, the odds of having a girl are now somehow higher. We know that a baby’s sex is determined at the point of conception by its chromosomal makeup. The sex of a baby is not determined by other children in the same family. To try to have another child because you think it could be a girl or boy is a case of the Gambler’s Fallacy error in logic.

Calculate the standard deviation of the numbers1 and numbers2 lists and compare the differences. 

```python
# A list of numbers - four small and one big.
numbers1 = np.array([1, 1, 1, 1, 10])

# A list of numbers - all close to each other.
numbers2 = np.array([2, 2, 3, 3, 4])
```

```python
# Numbers1 mean.
numbers1_m = np.mean(numbers1)

# The full calculation standard deviation calculation.
numbers1_sd = np.sqrt(np.mean(np.square(numbers1 - np.mean(numbers1))))

# Numbers2 mean.
numbers2_m = np.mean(numbers2)

# The full calculation standard deviation calculation.
numbers2_sd = np.sqrt(np.mean(np.square(numbers2 - np.mean(numbers2))))
```

```python
print(f'Numbers1 = {numbers1} mean = {numbers1_m} standard deviation = {numbers1_sd}')
```

```python
print(f'Numbers2 = {numbers2} mean = {numbers2_m} standard deviation = {round(numbers2_sd,3)}')
```

```python
# Subtract the mean from each of the values.
diff1 = numbers1 - numbers1.mean()
diff2 = numbers2 - numbers2.mean()

# Create a plot.
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the diff1 and diff2 array, each value spaced out evenly along the x axis.
ax.plot(range(len(diff1)), diff1, 'blue', marker='.', linestyle='none', label="Numbers1: Distance From Mean")
ax.plot(range(len(diff2)), diff2, 'r.', label="Numbers2: Distance From Mean")

# Plot the squared of the diff1 and diff2 array, each value spaced out evenly along the x axis.
ax.plot(range(len(diff1)), np.square(diff1), color='blue', marker='*', linestyle='none', label="Numbers1: $(DistanceFromMean)^2$")
ax.plot(range(len(diff2)), np.square(diff2), color='red', marker='*', linestyle='none', label="Numbers2: $(DistanceFromMean)^2$")

# Plot the legend
plt.legend()

# Plot the y=0 line.
ax.axhline(y=0.0, color='grey', linestyle='-');
```

Show that the difference between the standard deviation calculations is greatest for small sample sizes.

**Standard Deviation Formula** population standard deviation

$$\sigma = \sqrt{\frac{{\sum_{}(x_i - \mu)^2}}{N}}$$

$\sigma$ = population standard deviation<br>
N = the size of the population<br>
$x_i$ = each value in the population<br>
$\overline{x}$ = the population mean<br>

Source: <a href="https://www.google.com/search?q=standard+deviation+formula&oq=Standard+Deviation+Formula&aqs=chrome.0.35i39l2j0i512l8.839j0j7&sourceid=chrome&ie=UTF-8" target="_blank">Google: Standard Deviation Formula</a>

```python
# Population standard deviation.
sd_pop_fun = lambda sample: np.sqrt(np.mean(np.square(sample - np.mean(sample))))
```

**Bessel's correction** - sample standard deviation

$$s = \sqrt{\frac{{\sum_{}(x_i - \overline{x})^2}}{n-1}}$$

$s$ = sample standard deviation<br>
n = the number of observations<br>
$x_i$ = the value of one observation<br>
$\overline{x}$ = the mean value of all observations<br>

<a href="https://www.statisticshowto.com/bessels-correction/" target="_blank">Statistics How To: Bessel’s Correction</a>

```python
# Sample standard deviation.
def sd_sample_fun(sample):
    numerator = sum(np.square(sample - np.mean(sample)))
    denominator = len(sample) - 1
    result = np.sqrt(numerator / denominator)
    return result
```

```python
# Create 100000 samples of size 5 - standard deviation is 2.0.
samples5 = np.random.normal(0.0, 2.0, (100000, 5))

# Calculate standard deviations for 100000 samples of size 5
sd_pop_5 = [sd_pop_fun(s) for s in samples5]

# Calculate standard deviations for 100000 samples of size 5
sd_sample_5 = [sd_sample_fun(s) for s in samples5]
```

```python
# View a histogram - hopefully we can see the estimate is too small.
fig, ax = plt.subplots(figsize=(12, 6))

# Plot histograms.
plt.hist(sd_pop_5, bins=100, color='b', alpha=0.25, label="Population")
plt.hist(sd_sample_5, bins=100, color='g', alpha=0.25, label="Sample")

# Add legend 
plt.legend()

# Add title
plt.title("Population Vs. Sample Deviation n = 5")

# Draw a vertical line where the actual standard deviation is.
plt.axvline(x=2.0, color='r');
```
